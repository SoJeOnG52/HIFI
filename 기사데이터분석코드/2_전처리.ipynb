## 01. xlsx 파일 불러오기

```python
import numpy as np
import pandas as pd
from konlpy.tag import Okt
```

```python
df1 = pd.read_excel('./예적금_금융상품종류_유사도매핑_ train.xlsx')
df2 = pd.read_excel('./예적금_금융상품종류_유사도매핑_ val.xlsx')
```

## 02. 한글 외 문자 제거

```python
# 정규 표현식을 이용해서 한글 외의 문자는 전부 제거
df1['page1'] = df1['page1'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","") 
df1['page2'] = df1['page2'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")

df2['page1'] = df2['page1'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
df2['page2'] = df2['page2'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
```

## 04. 형태소 분석 후 morrphs열에 추가

### 1) df1(train)

```python
okt = Okt()
df1['page1_morphs'] = ""
df1['page2_morphs'] = ""

for index, row in df1.iterrows():
    text = row['page1']  # df1의 'page1' 열 값
    page1_morphs = okt.morphs(text)  # 형태소 분석 수행
    df1.at[index, 'page1_morphs'] = str(page1_morphs)  # 분석 결과를 'page1_morphs' 열에 저장
    

for index, row in df1.iterrows():
    text = row['page2']  # df1의 'page2' 열 값
    page2_morphs = okt.morphs(text)  # 형태소 분석 수행
    df1.at[index, 'page2_morphs'] = str(page2_morphs)  # 분석 결과를 'page1_morphs' 열에 저장
    
df1
```

### 2) df2(val)

```python
okt = Okt()
df2['page1_morphs'] = ""
df2['page2_morphs'] = ""


for index, row in df2.iterrows():
    text = row['page1']  # df2의 'page1' 열 값
    page1_morphs = okt.morphs(text)  # 형태소 분석 수행
    df2.at[index, 'page1_morphs'] = str(page1_morphs)  # 분석 결과를 'page2_morphs' 열에 저장
    
# 각 행별로 형태소 분석 수행하여 저장
for index, row in df2.iterrows():
    text = row['page2']  # df2의 'page2' 열 값
    page2_morphs = okt.morphs(text)  # 형태소 분석 수행
    df2.at[index, 'page2_morphs'] = str(page2_morphs)  # 분석 결과를 'page2_morphs' 열에 저장
    
df2
```

## 05. 불용어 리스트(korean.txt)

```python
with open('./korean.txt', 'r', encoding='utf-8') as file:
    stopwords_list = [line.strip() for line in file]
```

```python
stopwords_list
```

## 06. 불용어 및 1글자 제거 

### 1) df1

```python
# 형태소 분석 결과에서 불용어 및 한 글자 이하 단어 제거하여 저장
df1['page1_morphs'] = df1['page1'].apply(lambda x: [word for word in okt.morphs(x) if word not in stopwords_list and len(word) > 1])
df1['page2_morphs'] = df1['page2'].apply(lambda x: [word for word in okt.morphs(x) if word not in stopwords_list and len(word) > 1])

# 결과 출력
print(df1['page1_morphs'])
print(df1['page2_morphs'])
```

### 2) df2

```python
# 형태소 분석 결과에서 불용어 및 한 글자 이하 단어 제거하여 저장
df2['page1_morphs'] = df2['page1'].apply(lambda x: [word for word in okt.morphs(x) if word not in stopwords_list and len(word) > 1])
df2['page2_morphs'] = df2['page2'].apply(lambda x: [word for word in okt.morphs(x) if word not in stopwords_list and len(word) > 1])

# 결과 출력
print(df2['page1_morphs'])
print(df2['page2_morphs'])
```
